## 1. Teoria dos Conjuntos e Conceitos Básicos
Para a formulação matemática da probabilidade, utilizamos a teoria dos conjuntos.

- **União ($A \cup B$):** O evento ocorre se $A$ ou $B$ (ou ambos) ocorrem.
- **Intersecção ($A \cap B$):** O evento ocorre se $A$ e $B$ ocorrem simultaneamente.
- **Complementar ($\overline{A}$ ou $A^c$):** O evento ocorre se $A$ não ocorre.

**Eventos Mutuamente Exclusivos:** Diz-se que os eventos $A$ e $B$ são mutuamente exclusivos se $A \cap B = \emptyset$.

## 2. Experimento Aleatório

Um experimento aleatório é um experimento que pode ser
repetido inúmeras vezes sob as mesmas condições, sendo o
seu resultado incerto.

- *Lançamento de um dado*: quando você joga um dado, obtém uma sequência de valores

    - X = [1,4,6,5,2,4,5,3,2,2]

- Não é possível saber qual será o próximo valor. Mas podemos, determinar a probabilidade de cada resultado possível


**Espaço amostral ($\Omega$)**: Representa o conjunto de todas as saídas possíveis de um experimento aleatório. (Em uma moeda, seria {cara, coroa})


**Evento($A$)**: é um subconjunto do espoaço amostral. (Em uma moeda, seria {coroa})


**Evento impossível($\emptyset$)**: é o evento que nunca ocorre. (Em um dado, $\emptyset = \{x \in \Omega \mid x > 6\}$)

## 3. Axiomas da probabilidades (Axiomas de Kolmogorov)

**Definição**: Sejam $\Omega$ o espaço amostral e A um evento em $\Omega$.
Então, Uma função $P(\cdot)$ é probabilidade se satisfaz:
1. $0 \leq P(A) \leq 1$, para todo $A \subset \Omega$ (retornar um valor entre 0 e 1 para qualquer A contido no $\Omega$)
2. $P(\Omega) = 1$ (sempre vai ser igual a 1)
3. Se $A_1, A_2, \dots$ são mutuamente exclusivos, $P(\cup_{i=1}^{\infty} A_i) = \sum_{i=1}^{\infty} P(A_i)$ (minha probabilidade é a soma dos valores)



### Resumo direto

- Probabilidade não-negativa: A probabilidade de qualquer evento é sempre um número entre 0 e 1, o que é lógico, pois não podemos ter probabilidades negativas ou maiores que 100%.


- Probabilidade do espaço amostral: A probabilidade de que algo (qualquer resultado possível) aconteça é 1 (ou 100%).


- Aditividade para eventos mutuamente exclusivos: Se você tem eventos que não podem ocorrer ao mesmo tempo (mutuamente exclusivos), a probabilidade de que qualquer um deles ocorra é a soma das probabilidades individuais de cada evento.

  - A probabilidade de obter um número par OU um número ímpar é a soma das probabilidades individuais: $P(A \cup B) = P(A) + P(B) = \frac{1}{2} + \frac{1}{2} = 1$

### 3.1 Probabilidade Clássica

**Definição:** Se um experimento aleatório tiver $n(\Omega)$ resultados mutuamente exclusivos e igualmente possíveis, e se um evento $A$ contiver $n(A)$ desses resultados, a probabilidade de ocorrência desse evento é definida por:

$P(A) = \frac{n(A)}{n(\Omega)} = \frac{|A|}{|\Omega|}$.

```python
"""
Uma urna contém fichas numeradas de 1 a 20.
Supondo que alguém escolha uma dessas fichas ao acaso,
qual a probabilidade de que a ficha escolhida contenha um número maior que 9?
"""
```

```python
import random

n = 20 # espaço amostral
urna = range(1,n+1)
valor = 9 # a probabilidade de alguém pegar 9
f = 0 # contador de quantas vezes eu vou retirar uam bola
ns = 100 # numero de simulaçoes
for i in range(0, ns):
  ficha = random.choice(urna)
  if (ficha > valor):
    f = f + 1

print('Frequencia: ', f/ns)
print('Probabilidade exata: ', len(range(valor+1, n+1))/n)
```

## 4. Probabilidade frequentista

**Interpretação frequentista:** A probabilidade de um evento é igual à sua frequência de ocorrência em um grande número de experimentos:

$P(A) = \lim_{n\to\infty} \frac{n_A}{n}$

Sendo que $n_A$ é o número de vezes que o evento $A$ ocorre em $n$ experimentos.

```python
"""
A interepretação frequentista pode ser vista na prática através de simulação computacional. No código a seguir, consideramos o lançamento de uma moeda justa.
"""
```

```python
import random as random

ns = 100 # Numero de lançamentos
moeda = ['cara', 'coroa']
nc = 0 # Numero de caras
for n in range(0, ns):
  saida = random.choice(moeda)
  if (saida == 'cara'):
    nc += 1

print("Probabilidade de cara:", nc/ns)
print("Probabilidade de coroa:", (ns-nc)/ns)
```

```python
"""
Considere o problema proposto pelo matemático suíço Jakob Bernoulli (1654-1705). Uma urna contém 3.000 bolas brancas e 2.000 bolas pretas. Calcule a probabilidade de retirar uma bola preta da urna.
"""
```

```python
import random

urna = ['B']*3000 + ['P']*2000 # cria uma lista para a urna
nsim = 100
fp = 0
for i in range(0, nsim):
  saida = random.choice(urna)
  if saida == 'P':
    fp += 1

print("Probabilidade:", fp/nsim)
```

### Probabilidade da união de dois eventos

**Teorema**: Para dois eventos A e B em um mesmo espaço amostral $\Omega$

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

image.png

Sejam A e B eventos em um mesmo espaço amostral $\Omega$ e $\emptyset$ é o evento impossível. Seja ainda $A^c = \overline{A}$ o complementar de A.

Então:
* $P(\emptyset) = 0$,
* $P(A) = 1 - P(A^c)$,
* Se $A \subset B$, então $P(A) \leq P(B)$.

Para provarmos esse teorema, temos:

$A \cup \emptyset = A$

$P(A \cup \emptyset) = P(A)$

$P(A) + P(\emptyset) - P(A \cap \emptyset) = P(A)$

$P(A) + P(\emptyset) = P(A)$

$P(\emptyset) = 0$

pois $P(A \cap \emptyset) = 0$,
uma vez que A e $\emptyset$ nunca ocorrem ao mesmo tempo ($A \cap \emptyset = \emptyset$), pois $\emptyset$ é o evento impossível.

No caso da segunda propriedade:
$A \cup A^c = \Omega \Rightarrow P(A \cup A^c) = P(\Omega) \Rightarrow P(A) = 1 - P(A^c)$
pois $P(\Omega) = 1$ e $P(A \cup A^c) = P(A) + P(A^c)$.

Finalmente, temos que se $A \subset B$, então, $B = A \cup (A^c \cap B)$.
Logo, $P(B) = P(A) + P(A^c \cap B)$.
Assim, como $P(A^c \cap B) \geq 0$, então $P(A) \leq P(B)$.

Podemos calcular a frequência de ocorrência de caras:

```python
import numpy as np

n = 10
n_caras = 0
moeda = ['cara', 'coroa']
for i in range(0, n):
  face = random.choice(moeda)
  if face == 'cara':
    n_caras += 1

print('Frequêcia:', n_caras/n)
```

Variando o número de simulações (lançamentos)

```python
import numpy as np
import matplotlib.pyplot as plt

moeda = ['cara', 'coroa']
p = 0.5
lcaras = []
ln = []
for n in range(1, 1000):
  n_caras = 0
  for i in range(0, n):
    face = random.choice(moeda)
    if face == 'cara':
      n_caras += 1
  fa = n_caras/n
  lcaras.append(fa)
  ln.append(n)

plt.plot(ln, lcaras)
plt.axhline(y=p, color='r', linestyle='-')
plt.xlabel('Número de lançamentos')
plt.ylabel('Frequencia de caras')
plt.show()
```

Da mesma forma, podemos simular o lançamento de um dado

```python
dado = [1,2,3,4,5,6]

n = 1000
vfaces = np.zeros(6)
for i in range(0, n):
  face = np.random.choice(dado)
  vfaces[face-1] = vfaces[face-1] + 1

print('Resultado total das jogads:', vfaces)
print('Valor Teorico:', round(1/len(dado),2))
print('Frequencias obtidas:', vfaces/n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

moeda = ['cara','cara', 'coroa','coroa','coroa','coroa','coroa','coroa','coroa','coroa']
p = 0
for m in moeda:
  if m == 'cara':
    p = p + 1
p = p/len(moeda)
print('p = ', p)

lcaras = []
ln = []
for n in range(1,1000):
  ncaras = 0
  for i in range(0,n):
    face = np.random.choice(moeda)
    if face == 'cara':
      ncaras = ncaras + 1
  fa = ncaras/n
  lcaras.append(fa)
  ln.append(n)
plt.plot(ln,lcaras)
plt.axhline(y=p, color='r', linestyle='-')
plt.xlabel('Número de lançamentos')
plt.ylabel('Frequência de caras')
plt.show()
```

Exemplo: Uma urna contém fichas numeradas de 1 a 20. Supondo que alguém escolha uma dessas fichas ao acaso, qual é a probabilidade de que a ficha escolhida contenha um número maior do que 9?

```python
import random as random

nbolas = 20 # numero de bolas
urna = range(1, nbolas + 1)
fa = 0 # frequencia de sucessos (numero > 9)
ns = 100 # numero de simulacoes
for n in range(0, ns): # para ns lancamentos
  saida = random.choice(urna)
  if saida > 9:
    fa += 1

print('Frequencia:', fa/ns)
```

Outra maneira de simular

```python
import random
import numpy as np

n = 20
urna = np.arange(1,n+1)
valor = 9
f = 0 # contador
ns = 100
for i in range(0,ns):
  ficha = random.choice(urna)
  if(ficha > valor):
    f = f + 1
print('Frequencia:', f/ns)
print('Probabilidade exata:', len(range(valor+1, n+1))/n)
```
