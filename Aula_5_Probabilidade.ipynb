### Valor Esperado (Esperança Matemática)

**Definição:** Seja $X$ uma variável aleatória discreta com distribuição de probabilidade $P(X = x)$. O valor esperado (médio) ou esperança matemática de $X$ é definido por:

$$E[X] = \sum_{i=1}^{\infty} x_i P(X = x_i)$$

```python
from random import randint
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

medias = []
S = range(1, 1000) # número de simulaçoes
for s in S: # Simula
    saidas_dado = []
    for n in range(0,s):
        saidas_dado.append(randint(1,6))

    medias.append(np.mean(saidas_dado))

plt.figure(figsize=(8,6))
```

**Exemplo:** Uma padaria produz pães que proporcionam um lucro de R\\$1,00 real, quando vendidos, ou um prejuízo de R\$0,50 quando são encaminhados para virarem farinha, no outro dia.  Se a probabilidade de vender um pão é igual a 0,8, qual o lucro médio da padaria?


```python
import numpy as np

Ns = 1000 # número de simulacoes
p = 0.8 # probabilidade de vender um pão
lucro = 1 # lucro por pao
prejuizo = 0.5 # prejuizo por pao
ganho = 0 # lucro total acumulado

for s in range(0, Ns):
    # np.random.uniform() gera um valor entre 0 e 1
    if np.random.uniform() < p:
        ganho += lucro
    else:
        ganho -= prejuizo

print('Lucro médio simulado:', round(ganho/Ns, 2))

# O valor teórico esperado seria: (1 * 0.8) - (0.5 * 0.2) = 0.8 - 0.1 = 0.7
print('Lucro médio teórico:', round((lucro * p) - (prejuizo * (1 - p)), 2))
```

### Esperança de uma Variável Aleatória Contínua

**Definição:** Seja $X$ uma variável aleatória contínua com função densidade de probabilidade $f(x)$. O valor esperado (médio) ou esperança matemática de $X$ é definido por:

$$E[X] = \int_{-\infty}^{\infty} x f(x) dx$$

### Valor Esperado de uma Função de uma Variável Aleatória

**Definição:** Seja $g(X)$ uma função da variável aleatória discreta $X$. Então, o valor esperado de $g(X)$ é definido por:

$$E[g(X)] = \sum_{i=1}^{\infty} g(x_i) P(X = x_i)$$

### Valor Esperado de uma Função de uma Variável Aleatória Contínua

**Definição:** Seja $g(X)$ uma função da variável aleatória contínua $X$ com função de densidade de probabilidade $f(x)$. Então, o valor esperado de $g(X)$ é definido por:

$$E[g(X)] = \int_{-\infty}^{\infty} g(x) f(x) dx$$

### Propriedades da Esperança

**Teorema:** Seja $X$ uma variável aleatória.

* Se $X = c$, onde $c$ é uma constante, então:
$$E[X] = E[c] = c$$

* Seja $c$ uma constante, então:
$$E[cX] = cE[X]$$

* O valor esperado é uma função linear. Sejam $a$ e $b$ constantes, então:
$$E[aX + b] = aE[X] + b$$

### Momentos estatístico

**Definição:** Seja $X$ uma variável aleatória discreta com valores $x_1, x_2, \dots, x_k$. O momento de ordem $n$ de $X$ é definido por:

$$E[X^n] = \sum_{i=1}^{k} x_i^n P(X = x_i)$$

Para o caso contínuo, o momento de ordem $n$ é definido por:

$$E[X^n] = \int_{-\infty}^{\infty} x^n f(x) dx$$

**Definição:** Seja $X$ uma variável aleatória contínua com função densidade de probabilidade $f(x)$. O momento estatístico de ordem $n$ de $X$ é definido por:

$$E[X^n] = \int_{-\infty}^{\infty} x^n f(x) dx$$

### Momento Central

**Definição:** Seja $X$ uma variável aleatória.

*   Se $X$ é uma variável aleatória discreta, o momento central de ordem $n$ ($n > 0$) de $X$ é definido por:

$$\mu_n = E[(X - E[X])^n] = \sum_{i} (x_i - E[X])^n P(X = x_i)$$

*   Se $X$ é uma variável aleatória contínua, o momento central de ordem $n$ ($n > 0$) de $X$ é definido por:

$$\mu_n = E[(X - E[X])^n] = \int_{-\infty}^{\infty} (x - E[X])^n f(x) dx$$

### Variância de uma Variável Aleatória

**Definição:** A variância da variável aleatória $X$ é o segundo momento central, definida por:

$$V(X) = \sigma^2 = E[(X - E[X])^2]$$

O **desvio padrão** é a raiz quadrada da variância:

$$\sigma = \sqrt{V(X)}$$

**Teorema:** Seja $X$ uma variável aleatória com valor esperado $E[X]$. Então,

$$V(X) = E[X^2] - (E[X])^2$$

### Variância de uma Função de uma Variável Aleatória

**Definição:** Seja $g(X)$ uma função da variável aleatória $X$. Então,

$$V[g(X)] = E[g(X)^2] - (E[g(X)])^2$$

### Propriedades da Variância

**Teorema:** Seja $X$ uma variável aleatória e $a$ e $b$ constantes. Então,

$$V(aX + b) = a^2V(X)$$
