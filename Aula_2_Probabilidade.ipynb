## Probabilidade Condicional

Sejam dois eventos A e B em um mesmo espaço amostral Ω. A probabilidade condicional de A dado que B ocorreu é definida por:

P(A|B) = P(A ∩ B) / P(B)

P(B) > 0.

É comum também escrevermos

P(A ∩ B) = P(A, B).


### Resumo — Probabilidade Condicional

Considere um experimento aleatório com espaço amostral Ω e um resultado elementar ω ∈ Ω. Ao observarmos um evento B ⊆ Ω, a probabilidade dos resultados é atualizada de acordo com essa informação.

- Se ω ∈ B, então a probabilidade condicional \(P(\omega \mid B)\) é proporcional à probabilidade original \(P(\omega)\), pois a observação de B afeta a chance de ocorrência de ω.
- Se ω ∉ B, então \(P(\omega \mid B) = 0\), pois ω é incompatível com a ocorrência de B.

Como a soma das probabilidades condicionais deve ser 1, obtemos um fator de normalização que leva à expressão:


$$
P(\omega \mid B) =
\begin{cases}
\dfrac{P(\omega)}{P(B)}, & \text{se } \omega \in B \\
0, & \text{se } \omega \notin B
\end{cases}
$$

Esse raciocínio fundamenta a definição de probabilidade condicional e está na base do Teorema de Bayes.


## Eventos independentes

Teorema: A e B são eventos independentes se, e somente se,

P(A ∩ B) = P(A)P(B).

Definição: Os eventos $B_1, B_2, \ldots, B_n$ formam uma partição do espaço amostral se:

* $B_i \cap B_j = \emptyset$, para $i \neq j$
* $\bigcup_{i=1}^{n} B_i = \Omega$
* $P(B_i) \ge 0$, para $i = 1, \ldots, n$.

Teorema: Seja A um evento no espaço amostral Ω e seja
$B_1, \ldots, B_n$ uma partição amostral de Ω. Podemos escrever A
considerando tal partição:
$$A = \bigcup_{i=1}^{n} (A \cap B_i)$$

Teorema: Sejam $B_1, B_2, \ldots, B_n$ uma partição do espaço amostral $Ω$. Então, qualquer evento $A \in Ω$ pode ser escrito como:
$$P(A) = \sum_{i=1}^{n} P(A|B_i)P(B_i)$$

### Teorema de Bayes

Teorema: Seja $B_1, B_2, \ldots, B_n$ uma partição do espaço amostral $\Omega$
e $A$ um evento em $\Omega$. Então


$$P(B_i|A) = \frac{P(A|B_i)P(B_i)}{\sum_{j=1}^{n} P(A|B_j)P(B_j)}, \quad i = 1, 2, \ldots, n.$$
A prova desse teorema considera o teorema da probabilidade total
e a probabilidade condicional. Ou seja, escrevemos:


$$P(B_i|A) = \frac{P(A \cap B_i)}{P(A)} = \frac{P(A|B_i)P(B_i)}{\sum_{j} P(A|B_j)P(B_j)}$$
