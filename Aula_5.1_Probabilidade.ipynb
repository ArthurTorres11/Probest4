## Função de Distribuição Acumulada (CDF)

**Definição Geral:**
A função distribuição acumulada, ou simplesmente função de distribuição, de uma variável aleatória $X$ é definida por:

$$F(x) = P(X \le x)$$

### 1. Variável Aleatória Discreta
Seja $X$ uma variável aleatória discreta. A função de distribuição é definida por:

$$F(x) = P(X \le x) = \sum_{x_i \le x} P(X = x_i)$$

### 2. Variável Aleatória Contínua
Seja $X$ uma variável aleatória contínua. A função de distribuição é definida por:

$$F(x) = P(X \le x) = \int_{-\infty}^{x} f(s)ds$$

### Propriedades da CDF

**Teorema:** A função de distribuição de uma variável aleatória $X$ satisfaz as seguintes propriedades:

*   $0 \le F(x) \le 1$
*   $F(x)$ é não decrescente
*   $\lim_{x \to -\infty} F(x) = 0$ e $\lim_{x \to \infty} F(x) = 1$

### Cálculos e Relações

**1. Probabilidade em um intervalo:**
Para qualquer variável aleatória $X$, a probabilidade de $X$ estar no intervalo $(a, b]$ é:

$$P(a < X \le b) = F(b) - F(a)$$

**2. Relação com a Função Densidade de Probabilidade (PDF):**
No caso de uma variável aleatória contínua, a densidade de probabilidade $f(x)$ é a derivada da função de distribuição acumulada:

$$f(x) = \frac{d}{dx}F(x)$$
